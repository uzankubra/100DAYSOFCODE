```python
# Gerekli kütüphaneleri yükleme
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Iris veri setini yükleme
iris = sns.load_dataset("iris")
```

```python
# ilk 5 satırını gösterme
print(iris.head())
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(iris['sepal_length'], kde=True, color='skyblue')
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Dağılım grafiği
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris, hue='species', palette='viridis', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Kutu grafiği
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=iris, palette='Set2')
plt.title('Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```

```python
# Keman grafiği
plt.figure(figsize=(12, 8))
sns.violinplot(x='species', y='petal_width', data=iris, inner='quartile', palette='pastel')
plt.title('Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.show()
```

```python
# Çiftli grafik
sns.pairplot(iris, hue='species', height=2.5)
plt.suptitle('Pair Plot of Iris Data', y=1.02)
plt.show()
```

```python
# Ortak grafik
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='kde', color='purple')
plt.suptitle('Joint Plot of Sepal Length and Sepal Width', y=1.02)
plt.show()
```

```python
# Facet Grid
g = sns.FacetGrid(iris, col='species', hue='species', height=5)
g.map(plt.scatter, 'sepal_length', 'sepal_width', alpha=0.7)
g.add_legend()
plt.suptitle('Scatter Plots of Sepal Dimensions by Species', y=1.02)
plt.show()
```
